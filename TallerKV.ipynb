### Resolución

Para el numeral 1, 2 y 3 es necesario conocer la aceleracion angular primero.
Para ello, se realiza una sumotaria de torques en $O$
$$\Sigma \vec{\tau_O}=\vec{r} \times \vec{F}$$
$$\Sigma \vec{\tau_O}=\frac{b}{2} (sen \theta \vec{i} + cos \theta \vec{j}) \times mg \vec{j}$$
$$\Sigma \vec{\tau_O}=\frac{b}{2} mg sen \theta \vec{k}$$
Conocemos que la el torque es igual a la derivada respecto al tiempo de la cantidad de movimiento, por tanto, se tiene:
$$\Sigma \vec{\tau_O}=I_zz \cdot \dot{w_z} \vec{k}$$
Donde $I_zz$ debe ser trabajado con teorema de ejes paralelos.
$$I_zz=\frac{1}{12} (b^{2}+b^{2}) + m D^{2}$$
Con $D=b/2$, se tiene:
$$I_zz=\frac{5}{12} m b^{2}$$
Luego, igualando las ecuaciones:
$$\dot{w}=\alpha=\frac{5}{12} \frac{g}{b} sen \theta$$
Se sabe:
$$\int_{0}^{w}\! w\,dw = \int_{0}^{\theta}\! \alpha\,d\theta$$
$$w^{2}=\frac{12}{5} \frac{g}{b}\int_{0}^{\theta}\! sen \theta\,d\theta$$
$$w^{2}=\frac{12}{5} \frac{g}{b} (1-cos \theta)$$

## Numeral 1.
Para conocer el ángulo en el que pierde contacto, consideramos $N=0$, de donde:
$$\theta=cos^{-1}(6/11)$$
$$\theta=56.9°$$
Que es el ángulo en el que pierde contacto.

```python
import numpy as np #importar libreria matematica
import plotly.express as px #herramienta de plot
import plotly.graph_objects as go
#Análisis
th=np.arange(0,60,0.1) #creando lista numérica
vth = np.array(th) #transformado a una matriz
w2=(2000*9.81/5)*(11*np.cos(th/180*np.pi)-6)
#plot
fig=px.line(x=th,y=w2,labels={'x':'Ángulo de inclinación tetha [°]','y':'Normal [N]'},title="Variación de la normal con respecto al ángulo de giro")
fig.show()
```

*Análisis de la gráfica. La ecuación planteada para describir la variación de la normal con respecto al ángulo de giro es útil unicamente para valores menores iguales 56.9° dado que en adelante, se debería realizar otro análisis. De hecho, este corte se da justamente cuando N=0. Además de ello, valores negativos para la normal, no tienen sentido.*

```python
import numpy as np #importar libreria matematica
import plotly.express as px #herramienta de plot
import plotly.graph_objects as go
#Análisis
th=np.arange(0,60,0.1) #creando lista numérica
vth = np.array(th) #transformado a una matriz
w2=(12/5)*9.81/3*(1-np.cos(th/180*np.pi))
#plot
fig=px.line(x=th,y=w2,labels={'x':'Ángulo de inclinación tetha [°]','y':'Velocidad angular [rad/s]'},title="Variación de la velocidad angular con respecto al ángulo de giro")
fig.show()
```
